```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize

text_data = ["Ошибка, критично!",
             "Штатная ситуация, ничего нового",
             "Отличная работа"]

tokenized_text = [word_tokenize(sentence.lower()) for sentence in text_data]

model = Word2Vec(sentences=tokenized_text, vector_size=100, window=5, min_count=1, sg=0)

vectorized_text = [np.mean([model.wv[word] for word in sentence], axis=0) for sentence in tokenized_text]

print(vectorized_text)
```

```python
for v in vectorized_text:
  print(model.wv.similar_by_vector(v, topn=1), end=" ")
```

```python
import pandas as pd

data = pd.read_csv("Titanic.csv")
data.head(10)
```

```python
data["Target"] = [
    "Молодой рабочий",
    "Светская состоятельная дама",
    "Молодая девушка",
    "Светская состоятельная дама",
    "Взрослый рабочий",
    "Мужчина",
    "Почтенный состоятельный господин",
    "Малыш",
    "Активная девушка",
    "Девушка-подросток из состоятельной семьи"
] + [None for el in data.iterrows()][10:]
data.head(10)
```

```python
train = data.head(10)
test = data.iloc[10:, :]
```

```python
train
```

```python
trainX = train.drop(columns=["Target"])
trainY = train.Target
trainX
```

```python
trainX = trainX.drop(columns=["Name", "Ticket", "Cabin"])
trainX.Sex = trainX.Sex.replace({"male": 0, "female": 1, })
trainX.Embarked = trainX.Embarked.replace({"S": 0, "C": 1, "Q": 2, })
trainX
```

```python
trainY
```

```python
def cosine_similarity(vec1, vec2):
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))

mv = model.get_word_vector("Почтенный состоятельный господин")
closest_word = None
max_similarity = -1

for word in model.get_words():
    word_vector = model.get_word_vector(word)
    similarity = cosine_similarity(mv, word_vector)
    if similarity > max_similarity:
        max_similarity = similarity
        closest_word = word

closest_word
```

```python
from catboost import CatBoostRegressor

# Initialize the CatBoost model
model_pred = CatBoostRegressor(iterations=100, depth=6, learning_rate=0.1, loss_function='MultiRMSE', eval_metric='MultiRMSE')

# Train the model
model_pred.fit(trainX, pd.DataFrame(trainVectors).reset_index(drop=True))
```
